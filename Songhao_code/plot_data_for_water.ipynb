```python
A = 3
def convert_string_to_complex(df):
    "convert a string in a DataFrame to Complex number"
    L = len(df["0"])
    for i in range(A):
        df[str(i)] = df[str(i)].astype(complex)
    return df
```

```python

state_pop = pd.read_csv("Diabatic_state_pop_for_h2o_state_0.csv")
state_pop = convert_string_to_complex(state_pop)
state_pop.head()
```

```python
plt.figure(figsize=(20, 12))
for i in range(A):
    plt.plot(state_pop["time"], state_pop[str(i)], label="state {:}".format(i), linewidth=5)
plt.tick_params(labelsize=60,\
               labelbottom=True,
               labelleft=True)
plt.legend(fontsize=40)
plt.xlabel("time(fs)", fontsize=60)
plt.ylabel("state population", fontsize=60)
plt.show()
```

```python
A = 3
state_pop_Z3 = convert_string_to_complex(pd.read_csv("Diabatic_state_pop_for_h2o_state_2_VECC_Z3.csv"))
state_pop_Z3.head()
```

```python
A = 3
state_pop_Z2 = convert_string_to_complex(pd.read_csv("Diabatic_state_pop_for_h2o_state_2_VECC_Z2.csv"))
state_pop_Z2.head()
```

```python
linestyle_list = ["solid", "dashed", "dashdot"]

plt.rcParams["mathtext.default"] = "regular"
plt.rc('text.latex', preamble=r'\usepackage[cm]{sfmath}')
f, (ax1, ax2, ax3) = plt.subplots(3,1,sharex=True,figsize=(20,30),
        constrained_layout=True)

# plot MCTDH state population
for i in range(A):
    term = term_symbol_dic[str(i)]
    ax1.plot(state_pop_mctdh_3["time(fs)"], state_pop_mctdh_3[str(i)], label="MCTDH, state {:} ({:})".format(i+1, term),\
             linestyle = linestyle_list[i], linewidth=6)
ax1.legend(loc="center left", fontsize=40)
ax1.tick_params(labelsize=60,\
               labelbottom=False,
               labelleft=True)

for i in range(A):
    term = term_symbol_dic[str(i)]
    ax2.plot(state_pop_Z2["time"], state_pop_Z2[str(i)], label="VECC T:S, Z:SD, state {:} ({:})".format(i+1, term), \
            linestyle = linestyle_list[i], linewidth=6)
ax2.legend(loc="center left",fontsize=40)
ax2.tick_params(labelsize=60,\
               labelbottom=False,
               labelleft=True)

for i in range(A):
    term = term_symbol_dic[str(i)]
    ax3.plot(state_pop_Z3["time"], state_pop_Z3[str(i)], label="VECC T:S, Z:SDT, state {:} ({:})".format(i+1, term), \
            linestyle = linestyle_list[i], linewidth=6)
ax3.legend(loc="center left", fontsize=40)
ax3.tick_params(labelsize=60,\
               labelbottom=True,
               labelleft=True)
f.supxlabel("t (fs)", fontsize=80)
f.supylabel("Population", fontsize=80)
plt.show();
```
