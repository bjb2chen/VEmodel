```python
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import numpy as np
```

# Plot 1d scan data over e mode 

```python
number_of_state = 3
```

```python
df=pd.read_json("nh3_1d_scan_data.json")
df
```

```python
plt.figure(figsize=(20,12))
for i in range(number_of_state):
#     adiabatic_state_name = "adiabatic_state_{:d}".format(i+1)
#     plt.plot(df["Grid"], df[adiabatic_state_name])
#     plt.scatter(df["Grid"], df[adiabatic_state_name], label=adiabatic_state_name, linewidth=5)
    diabatic_state_name = "diagonal_diabatic_state_{:d}".format(i+1)
    plt.scatter(df["Grid"], df[diabatic_state_name], label=diabatic_state_name, linewidth=5)
    plt.plot(df["Grid"], df[diabatic_state_name])
plt.legend()
plt.title("Diabatic potential (1d scan)", fontsize=40)
plt.show()
```

```python
plt.figure(figsize=(20,12))
for i in range(number_of_state):
    adiabatic_state_name = "adiabatic_state_{:d}".format(i+1)
    plt.plot(df["Grid"], df[adiabatic_state_name])
    plt.scatter(df["Grid"], df[adiabatic_state_name], label=adiabatic_state_name, linewidth=5)
plt.legend()
plt.title("Adiabatic potential (1d scan)", fontsize=40)
plt.show()
```

```python
plt.figure(figsize=(20,12))
for i in range(number_of_state):
    for j in range(i+1, number_of_state):
        name = "off_diagonal_diabatic_state_{:d}_{:d}".format(i+1, j+1)
        plt.scatter(df["Grid"], df[name], label=name, linewidth=5, alpha=.5)
        plt.plot(df["Grid"], df[name], linewidth=5, alpha=.5)
plt.legend()
plt.title("Diabatic off-diagonal potential", fontsize=40)
plt.show()
        
```

# Plot 2d scan data for e modes

```python
number_of_state = 3
```

```python
df=pd.read_json("nh3_2d_scan_data.json")
df.head()
```

```python
fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(111, projection='3d')

for i in range(number_of_state):
    name = "adiabatic_state_{:d}".format(i+1) 
    x = df["Grid_1"]
    y = df["Grid_2"]
    z = df[name]
    surf = ax.plot_trisurf(x, y, z, label=name, alpha=.5)
    surf._edgecolors2d = surf._edgecolor3d
    surf._facecolors2d = surf._facecolor3d

ax.legend(fontsize=20)
plt.title("Plot of adiabatic PES", fontsize=40)
plt.show()
```

```python
fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(111, projection='3d')

for i in range(number_of_state):
    name = "diagonal_diabatic_state_{:d}".format(i+1) 
    x = df["Grid_1"]
    y = df["Grid_2"]
    z = df[name]
    surf = ax.plot_trisurf(x, y, z, label=name, alpha=.5)
    surf._edgecolors2d = surf._edgecolor3d
    surf._facecolors2d = surf._facecolor3d

ax.legend(fontsize=20)
plt.title("Plot of diabatic PES", fontsize=40)
plt.show()
```
